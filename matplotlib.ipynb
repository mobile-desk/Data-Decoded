# Import

```python
import matplotlib.pyplot as plt
import numpy as np
```

# Simple Graph

```python
a = [1,2,3]
b = [2,4,6]
g = plt.plot(a,b)
```

```python
v = np.arange(0,10,0.5)
plt.plot(v, v**2)
```

# Customizing a graph

## Add title

```python
a = [1,2,3]
b = [2,4,6]
plt.plot(a,b)
plt.title("A Simple Graph")
```

## Label each axis

```python
a = [1,2,3]
b = [2,4,6]
plt.plot(a,b)

plt.xlabel('X')
plt.ylabel('Y')
```

## Customizing the ticks

```python
a = [1,2,3]
b = [2,4,6]
plt.plot(a,b)

plt.xticks([1.5, 2.5, 3.5, 4.5])
plt.yticks([1,2,3,4,5,6,7])
           
```

## Legend

```python
a = [1,2,3]
b = [2,4,6]
plt.plot(a,b, color='green', linewidth=10, marker='X', markersize=10, linestyle='--', label='f')
plt.legend()
plt.show()
```

## Resize graph

```python
a = [1,2,3]
b = [2,4,6]

plt.figure(figsize=(10,6))
plt.plot(a,b)
```

# Save Graph

```python
a = [1,2,3]
b = [2,4,6]
plt.plot(a,b)
plt.savefig('file.png', dpi=150)
```

# Bar Chart

```python
a = ['A', 'B', 'C']
b = [3,6,9]
```

```python
plt.bar(a,b)
```

## add patterns

```python
bars = plt.bar(a,b)

bars[0].set_hatch('/')
bars[1].set_hatch('o')
bars[2].set_hatch('*')

```

## Add colors

```python
bars = plt.bar(a,b)

bars[0].set_color('r')
bars[1].set_color('g')
bars[2].set_color('y')

```

# Bar vs Histogram

# Histogram

```python
np.random.seed(0)
data = np.random.normal(loc=0, scale=1, size=100)
data
```

```python

plt.hist(data)
```

## To specify the bins

```python
plt.hist(data, bins= 30, edgecolor='black')
plt.ylabel('no of people in a bin')
```

## Customize bin range

```python
bin_ranges = [-3, -2, -1, 0, 1, 2, 3]
plt.hist(data, bins= bin_ranges, edgecolor='black')
plt.ylabel('no of people in a bin')
```

# Scatter Plot

```python
np.random.seed(0)
x = np.random.rand(50)
y = 2 * x + np.random.normal(0, 0.1, 50)
```

```python
plt.scatter(x,y)
```

## Arguments it takes

### Size

```python
plt.scatter(x,y, s=100)
```

```python
# If you want the sizes to be different depending on a variable
plt.scatter(x,y, s=y)
```

### Color

```python
plt.scatter(x,y, c='green')
```

```python
#  custokmise the color based on an argument

plt.scatter(x,y, c=y)
```

```python
# use a single color palette

plt.scatter(x,y,c=y,cmap='Greens')
```

```python
# add a color label
plt.scatter(x,y,c=y,cmap='Greens')
cbar = plt.colorbar()
cbar.set_label('Highest')
```
