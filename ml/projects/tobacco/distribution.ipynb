# Import

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

# load data

```python
df = pd.read_csv('cleaned_tobacco.csv')
```

# Plot a histogram

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Cigarette Use Prevalence % (Focus group)'], kde=True, color='blue')
plt.xlabel('Cigarette Use Prevalence % (Focus group)')
plt.ylabel('Frequency')
plt.title('Histogram of Cigarette Use Prevalence % (Focus group)')
plt.show()
```

# Plot a density plot

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(df['Cigarette Use Prevalence % (Focus group)'], color='green', fill=True)
plt.xlabel('Cigarette Use Prevalence % (Focus group)')
plt.ylabel('Density')
plt.title('Density Plot of Cigarette Use Prevalence % (Focus group)')
plt.show()
```
