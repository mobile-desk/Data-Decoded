# Import

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, StandardScaler, PolynomialFeatures
from sklearn import linear_model #build linear regressioon model
from sklearn.metrics import mean_squared_error, r2_score #compute performance

#from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import DecisionTreeRegressor

from sklearn.svm import SVR

from sklearn.metrics import accuracy_score

import os
```

# Load Data

```python
df = pd.read_csv('cleaned_tobacco.csv')
df
```

# Create new dataset

```python

# Pivot the DataFrame using pivot_table
new_df = df.pivot_table(index=['Year', 'State', 'Cigarette Use Prevalence % (Focus group)'], columns='Demographic', values='Comparing (Focus group)', aggfunc='first')
```

```python
new_df.reset_index(inplace=True)
```

```python

# Create a folder named 'data' if it doesn't exist
if not os.path.exists('data'):
    os.makedirs('data')

# Get unique demographic values
demographic_values = df['Demographic'].unique()

# Iterate over each demographic value
for value in demographic_values:
    # Get rows where 'Demographic' column matches the current value
    rows_for_value = df[df['Demographic'] == value]
    
    # Select columns to include in the new dataset
    selected_columns = ['Year', 'State', 'Cigarette Use Prevalence % (Focus group)', 'Comparing (Focus group)']
    
    # Create a new DataFrame with selected columns
    new_df = rows_for_value[selected_columns]
    
    # Drop duplicate rows
    new_df = new_df.drop_duplicates()
    
    # Save the new DataFrame to a CSV file in the 'data' folder
    filename = f"data/{value}_dataset.csv"
    new_df.to_csv(filename, index=False)
```

## Split Data

```python
df.columns
```

```python
X = df[['Year', 'State', 'Demographic', 'Comparing (Focus group)']]
y = df['Cigarette Use Prevalence % (Focus group)']
```

```python
#split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

## Feature Scalling (Standardization/normalization)

### Train

```python
scaler = StandardScaler().fit(X_train)
```

```python
#transform
X_train_scaled = scaler.transform(X_train)
X_train_scaled
```

```python
X_train_scaled.mean(axis=0)
#its mean should be 0 if its standardized
```

```python
X_train_scaled.std(axis=0)
#std should be 1
```

### Test

```python
scaler1 = StandardScaler().fit(X_test)
```

```python
#transform
X_test_scaled = scaler1.transform(X_test)
X_test_scaled
```

```python
X_test_scaled.mean(axis=0)
#its mean should be 0 if its standardized
```

```python
X_test_scaled.std(axis=0)
#std should be 1
```

# Simple Linear Regression

## Build Linear Regression

```python
model = linear_model.LinearRegression()
```

## Build Training Model

```python
model.fit(X_train_scaled, y_train)
```

## Make Predictions

```python
y_pred = model.predict(X_test_scaled)
```

## Analysis

```python
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

accuracy = model.score(y_test, y_pred)
print(f'Accuracy, {accuracy} ')


print(f'Coefficients, {model.coef_}')
print(f'Intercept, {model.intercept_}')
print(f'Mean Squared Error {mse:.2f}')
print(f'Coefficient of determination {r2:.2f}')
```

# Polynomial Regression

```python

poly_features = PolynomialFeatures(degree=2, include_bias=False)
```

```python


# degree=2 means that we want to work with a 2nd degree polynomial: y = ß0 + ß1x + ß2x^2

# include_bias=False should be set to False, because we’ll use PolynomialFeatures together with LinearRegression() later on
```

```python
X_poly = poly_features.fit_transform(X)
```

```python
# Fit linear regression model
lin_reg = linear_model.LinearRegression()
lin_reg.fit(X_poly, y)
```

```python
# Predicting on the training data
y_pred = lin_reg.predict(X_poly)
```

## Analysis

```python
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)


accuracy = accuracy_score(y, y_pred)
print(f'Accuracy, {accuracy} ')

# Print model coefficients, MSE, and R^2 score
print(f'Coefficients: {lin_reg.coef_}')
print(f'Intercept: {lin_reg.intercept_}')
print(f'Mean Squared Error: {mse:.2f}')
print(f'Coefficient of determination (R^2): {r2:.2f}')
```
