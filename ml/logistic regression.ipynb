# Install Modules

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, StandardScaler
```

# Load Dataset

```python
df = pd.read_csv('framingham_heart_disease.csv')
df
```

```python
(df.isnull().sum()/4238)*100
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
X = df[['male', 'age', 'education', 'currentSmoker', 'cigsPerDay', 'BPMeds', 'prevalentStroke', 'prevalentHyp', 'diabetes', 'totChol', 'sysBP', 'diaBP', 'BMI', 'heartRate', 'glucose']]
y = df['TenYearCHD']
```

# Visualize Data

```python
X.shape
```

```python
y.shape
```

```python
# Concatenate features and target into a single DataFrame
df = pd.concat([X, y], axis=1)

# Visualize each feature against the target variable using scatter plots
for column in X.columns:
    plt.figure(figsize=(8, 6))
    sns.scatterplot(data=df, x=column, y='TenYearCHD', hue='TenYearCHD', palette='coolwarm', marker='o', s=100)
    plt.title(f'Scatter Plot of {column} vs. TenYearCHD')
    plt.xlabel(column)
    plt.ylabel('TenYearCHD')
    plt.legend(title='TenYearCHD', loc='upper right', labels=['No CHD', 'CHD'])
    plt.grid(True)
    plt.show()

#this should give the s(sigmoid) shape
```

# SPlit Data

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=1)
```

# Feature Scalling (Standardization/normalization)

## Train

```python
scaler = StandardScaler().fit(x_train)
```

```python
scaler.mean_
#mean for every column under the dataset
```

```python
# get the varience
scaler.scale_
```

```python
#transform
x_train_scaled = scaler.transform(x_train)
x_train_scaled
```

```python
x_train_scaled.mean(axis=0)
#its mean should be 0 if its standardized
```

```python
x_train_scaled.std(axis=0)
#std should be 1
```

## Test

```python
scaler1 = StandardScaler().fit(x_test)
```

```python
scaler1.mean_
#mean for every column under the dataset
```

```python
# get the varience
scaler1.scale_
```

```python
#transform
x_test_scaled = scaler1.transform(x_test)
x_test_scaled
```

```python
x_test_scaled.mean(axis=0)
#its mean should be 0 if its standardized
```

```python
x_test_scaled.std(axis=0)
#std should be 1
```

# Logistic Regression

```python
log_reg = LogisticRegression()
log_reg.fit(x_train_scaled, y_train)
print(log_reg)
```

# Make Predictions

```python
y_pred = log_reg.predict(x_test_scaled)
```

```python
y_test.shape
```

```python
y_pred.shape
```

# Confusion Matrix

```python
# you have TP(true positive), FP(false positive), FN(false negative), TN(true negative)
# anything false should be low


# ROC(receiver operating characteristics) curve

# Show Coefficient and Intercept
print(f'Coefficient: {log_reg.coef_}')
print(f'Intercept: {log_reg.intercept_}')
```

```python
confusion_matrix(y_test, y_pred)
```

# Eval;uate model

## aCCURACY

```python
(TP + TN) / (TP + TN + FP + FN) # maximum possible is 1.0
```

## Precision

```python
TP / (TP + FP)
```

## Recall

```python
TP / (TP + FN)
```

## F1 Score

```python
(2 * TP) / (2 * (TP + FP + FN))
```
