# Import

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

# Import data sets

```python
sns.get_dataset_names()
```

```python
x = sns.load_dataset("tips")
```

```python
x
```

# Scatter Plot

```python
sns.scatterplot(x='total_bill', y='tip', data=x, hue="sex", size='size')
```

# Regression Plot

```python
sns.regplot(x='total_bill', y='tip', data=x)
```

# Histogram

```python
sns.histplot(x['tip'], kde=True, bins=15)
```

# Bar Plots

```python
sns.barplot(x='day', y='tip', data=x)
```

# Box Plot

```python
sns.boxplot(x='day', y='tip', data=x, hue="sex")
```

# Strip Plot

```python
sns.stripplot(x='day', y='tip', data=x, hue='sex', dodge=True) #dodge: if not set to true all the different hue would over lap so use this to seprate
```

# Join Plot

```python
sns.jointplot(x='tip', y='total_bill', data=x, kind='reg')
# numerical only
#kde, hex
```

# Pairplot

```python
sns.pairplot(x, hue='sex')
```

# Heat Map

```python
x.corr(numeric_only=True)
```

```python
sns.heatmap(x.corr(numeric_only=True), annot=True, cmap='coolwarm')
```
